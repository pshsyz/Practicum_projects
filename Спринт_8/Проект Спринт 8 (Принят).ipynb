### –ó–∞–¥–∞—á–∞ 7

–°–≥—Ä—É–ø–ø–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—è–º –∑–∞–≤–µ–¥–µ–Ω–∏–π –∏ –Ω–∞–π–¥–∏—Ç–µ —Ç–æ–ø-15 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö —Å–µ—Ç–µ–π –≤ –ú–æ—Å–∫–≤–µ. –î–ª—è –Ω–∏—Ö –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞. –ü–æ–¥ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å—é –ø–æ–Ω–∏–º–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π —ç—Ç–æ–π —Å–µ—Ç–∏ –≤ —Ä–µ–≥–∏–æ–Ω–µ. –ö –∫–∞–∫–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∑–∞–≤–µ–¥–µ–Ω–∏–π –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è? –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ–ø—Ä–æ–≤–æ–¥–∏—Ç–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–º–∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è–º–∏.

–ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ.

–†–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –±—É–¥–µ–º —Ç–æ–ª—å–∫–æ —Å–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è.

–°—á–∏—Ç–∞–µ–º –¥–ª—è –Ω–∏—Ö —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∞

```python
# –ù–∞–∫–∏–¥—ã–≤–∞–µ–º —Ñ–∏–ª—å—Ç—Ä –ø–æ —Ñ–ª–∞–≥—É –°–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è
chains = df_merge[df_merge['chain'] == True]

# –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ –∫–∞–∂–¥–æ–π —Å–µ—Ç–∏
category_counts = chains.groupby(['name', 'category']).size().reset_index(name='count')
# –°–æ—Ä—Ç–∏—Ä—É–µ–º –∏ –æ—Ç–±–∏—Ä–∞–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏—é —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º
main_categories = category_counts.sort_values('count', ascending=False).drop_duplicates('name')

# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º
chain_sum = chains.groupby('name').agg(count=('name', 'count'),avg_rating=('rating', 'mean')).reset_index()
# –î–æ–±–∞–≤–ª—è–µ–º –æ—Å–Ω–æ–≤–Ω—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é –∫ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–µ
top_chains = chain_sum.merge(main_categories[['name', 'category']], on='name', how='left')

# –¢–æ–ø-15 –ø–æ –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤—É –∑–∞–≤–µ–¥–µ–Ω–∏–π 
top_chains = top_chains.sort_values(by='count', ascending=False).head(15)
top_chains
```
